## Data loading

### Subtask:
Load the sales data from the Excel file "SalesData (3).xlsx" into a pandas DataFrame.


**Reasoning**:
I need to load the excel file into a pandas DataFrame and display the first 5 rows.



```python
import pandas as pd

try:
    sales_data = pd.read_excel('SalesData (3).xlsx')
    display(sales_data.head())
except FileNotFoundError:
    print("Error: 'SalesData (3).xlsx' not found. Please ensure the file exists in the current directory or provide the correct path.")
except Exception as e:
    print(f"An error occurred: {e}")
```

## Data exploration

### Subtask:
Explore the loaded sales data to understand its characteristics.


**Reasoning**:
Explore the characteristics of the sales data by checking its shape, data types, descriptive statistics, unique values of categorical columns, and missing values.



```python
# Check the shape of the DataFrame
print("Shape of the DataFrame:", sales_data.shape)

# Examine the data types of each column
print("\nData types of each column:\n", sales_data.dtypes)

# Generate descriptive statistics for numerical columns
print("\nDescriptive statistics for numerical columns:\n", sales_data.describe())

# Explore unique values and frequencies for categorical columns
categorical_cols = ['Region', 'Item', 'Sales Rep']
for col in categorical_cols:
    print(f"\nUnique values and frequencies for '{col}':")
    print("Unique values:", sales_data[col].unique())
    print("Value counts:\n", sales_data[col].value_counts())

# Check for missing values in each column
print("\nMissing values in each column:\n", sales_data.isnull().sum())
```

## Data analysis

### Subtask:
Analyze the sales data to identify key trends and patterns.


**Reasoning**:
Calculate the total sales, average sales per region, and total units sold; Analyze sales trends over time; Analyze sales distribution by region and by product category; Investigate the relationship between 'Units' and 'Total'; Analyze sales performance by sales representative; Summarize the findings.



```python
# Calculate total sales, average sales per region, and total units sold
total_sales = sales_data['Total'].sum()
average_sales_per_region = sales_data.groupby('Region')['Total'].mean()
total_units_sold = sales_data['Units'].sum()

# Analyze sales trends over time
sales_data['OrderDate'] = pd.to_datetime(sales_data['OrderDate'])
sales_data['MonthYear'] = sales_data['OrderDate'].dt.to_period('M')
monthly_sales = sales_data.groupby('MonthYear')['Total'].sum()

# Analyze sales distribution by region and item
sales_by_region = sales_data.groupby('Region')['Total'].agg(['sum', 'mean'])
sales_by_item = sales_data.groupby('Item')['Total'].agg(['sum', 'mean'])

# Investigate the relationship between 'Units' and 'Total'
correlation = sales_data['Units'].corr(sales_data['Total'])

# Analyze sales performance by sales representative
sales_by_rep = sales_data.groupby('Sales Rep')['Total'].agg(['sum', 'mean'])

# Summarize findings
summary_table = pd.DataFrame({
    'Metric': ['Total Sales', 'Average Sales per Region', 'Total Units Sold', 'Correlation (Units, Total)'],
    'Value': [total_sales, average_sales_per_region.mean(), total_units_sold, correlation]
})

display(summary_table)
display(sales_by_region)
display(sales_by_item)
display(sales_by_rep)
display(monthly_sales)
```

## Data visualization

### Subtask:
Visualize the key findings from the data analysis.


**Reasoning**:
Create the visualizations as requested by the instructions, including bar charts for sales by region, item, and representative, and a line chart for monthly sales.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Bar chart for sales by region (Total Sales and Average Sales)
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

sales_by_region[['sum']].plot(kind='bar', ax=axes[0], legend=False)
axes[0].set_title('Total Sales by Region')
axes[0].set_xlabel('Region')
axes[0].set_ylabel('Total Sales')
axes[0].tick_params(axis='x', rotation=0)

sales_by_region[['mean']].plot(kind='bar', ax=axes[1], legend=False, color='orange')
axes[1].set_title('Average Sales by Region')
axes[1].set_xlabel('Region')
axes[1].set_ylabel('Average Sales')
axes[1].tick_params(axis='x', rotation=0)

plt.tight_layout()
plt.show()

# 2. Line chart for monthly sales
plt.figure(figsize=(12, 6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trends')
plt.xlabel('Month-Year')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()

# 3. Bar chart for sales by item (Total Sales)
plt.figure(figsize=(10, 6))
sales_by_item['sum'].sort_values(ascending=False).plot(kind='bar')
plt.title('Total Sales by Item')
plt.xlabel('Item')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Bar chart for sales by representative (Total Sales), sorted
plt.figure(figsize=(12, 6))
sales_by_rep['sum'].sort_values(ascending=False).plot(kind='bar')
plt.title('Total Sales by Sales Representative')
plt.xlabel('Sales Representative')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
